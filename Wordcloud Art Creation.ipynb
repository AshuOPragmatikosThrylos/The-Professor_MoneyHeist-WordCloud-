```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
stopwords = set(STOPWORDS)


def grey_color_func(word, font_size, position, orientation, random_state=None,
                    **kwargs):
    return "hsl({}, {}%, {}%)".format(str(random.randint(0, 150)),str(random.randint(80, 100)), str(random.randint(40, 50)))

def show_wordcloud(data, title = None):
    
    alice_mask = np.array(Image.open("C:\\Users\\Ashutosh\\Pictures\\Professor.png"))
    
    wordcloud = WordCloud(
        font_path="C:\\Users\\Ashutosh\\Documents\\OnlineWebFonts_COM_94baf3bf7ec52c408a74dd58b8bca5de\\La Casa W03 Extra Bold\\abc1.ttf",
        background_color='white',
        stopwords=stopwords,
        max_words=400,
        max_font_size=40, 
        scale=4,
#         height=30,
#         width=100,
        mask=alice_mask,
        repeat=True,
        random_state=1 # chosen at random by flipping a coin; it was heads
    ).generate(str(data)).recolor(color_func=grey_color_func, random_state=3)

    fig = plt.figure(1, figsize=(12, 12))
    plt.axis('off')
    if title: 
        fig.suptitle(title, fontsize=20)
        fig.subplots_adjust(top=2.3)

    plt.imshow(wordcloud,
           interpolation="bilinear")
    plt.show()
#     plt.imsave("Hello.png",wordcloud)      #object too deep for desired array
    
data = "MoneyHeist LacasaDePapel Professor Tokyo Berlin Oslo Helsinki Denver Nairobi Moscow Palmero Rio Lisbon Stockholm Arturo Sierra Pierto Tamayo Chernobyl Cameroon Plan Suicide  RoyalMintofSpain BankofSpain Dali 2.4 billion AlisonParker Governor Gold"
show_wordcloud(data)
```
